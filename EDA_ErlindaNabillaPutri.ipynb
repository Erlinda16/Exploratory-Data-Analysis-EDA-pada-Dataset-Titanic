# Import library yang diperlukan

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv('train.csv')
```

```python
# Menampilkan informasi awal tentang dataset
print(df.info())
```

```python
df.isnull().sum()
```

# Menangani missing values

```python
# Menangani missing values
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop(columns='Cabin', inplace=True)
```

# Menangani data duplikat

```python

df.drop_duplicates(inplace=True)
```

# Visualisasi distribusi umur

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Distribusi Umur Penumpang')
plt.xlabel('Umur')
plt.ylabel('Jumlah')
plt.show()
```

# Visualisasi tingkat keselamatan berdasarkan jenis kelamin

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Tingkat Keselamatan Berdasarkan Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Jumlah')
plt.legend(title='Selamat', labels=['Tidak', 'Ya'])
plt.show()
```

# Visualisasi tingkat keselamatan berdasarkan kelas penumpang

```python

plt.figure(figsize=(10,6))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Tingkat Keselamatan Berdasarkan Kelas Penumpang')
plt.xlabel('Kelas')
plt.ylabel('Jumlah')
plt.legend(title='Selamat', labels=['Tidak', 'Ya'])
plt.show()
```
